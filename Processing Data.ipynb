Classifying Birds by the Features of their Song

First will import all the metadata and features for each speices of bird.

Then combine all the metadata into a pandas DataFrame named metadata and the features into a pandas DataFrame named features.

```python
import pandas as pd
metadata_yellowhammer = pd.read_csv(r'Data\\Metadata\\yellowhammermeta.csv')
metadata_whitethroat = pd.read_csv(r'Data\\Metadata\\whitethroatmeta.csv')
metadata_turtledove = pd.read_csv(r'Data\\Metadata\\turtledovemeta.csv')
metadata_treesparrow = pd.read_csv(r'Data\\Metadata\\treesparrowmeta.csv')
metadata_stockdove = pd.read_csv(r'Data\\Metadata\\stockdovemeta.csv')
metadata_starling = pd.read_csv(r'Data\\Metadata\\starlingmeta.csv')
metadata_skylark = pd.read_csv(r'Data\\Metadata\\skylarkmeta.csv')
metadata_linnet = pd.read_csv(r'Data\\Metadata\\linnetmeta.csv')
metadata_lapwing = pd.read_csv(r'Data\\Metadata\\lapwingmeta.csv')
metadata_greypartridge = pd.read_csv(r'Data\\Metadata\\greypartridgemeta.csv')
metadata_goldfinch = pd.read_csv(r'Data\\Metadata\\goldfinchmeta.csv')
metadata_cornbunting = pd.read_csv(r'Data\\Metadata\\cornbuntingmeta.csv')
```

```python
metadata_dfs = [metadata_yellowhammer,
               metadata_whitethroat,
               metadata_turtledove,
               metadata_treesparrow,
               metadata_stockdove,
               metadata_starling,
               metadata_skylark,
               metadata_linnet,
               metadata_lapwing,
               metadata_greypartridge,
               metadata_goldfinch,
               metadata_cornbunting]
```

```python
metadata = pd.concat(metadata_dfs, ignore_index=True)
```

```python
features_yellowhammer = pd.read_csv(r'Data\\Features_Final\\yellowhammer_features.csv')
features_whitethroat = pd.read_csv(r'Data\\Features_Final\\whitethroat_features.csv')
features_turtledove = pd.read_csv(r'Data\\Features_Final\\turtledove_features.csv')
features_treesparrow = pd.read_csv(r'Data\\Features_Final\\treesparrow_features.csv')
features_stockdove = pd.read_csv(r'Data\\Features_Final\\stockdove_features.csv')
features_starling = pd.read_csv(r'Data\\Features_Final\\starling_features.csv')
features_skylark = pd.read_csv(r'Data\\Features_Final\\skylark_features.csv')
features_linnet = pd.read_csv(r'Data\\Features_Final\\linnet_features.csv')
features_lapwing = pd.read_csv(r'Data\\Features_Final\\lapwing_features.csv')
features_greypartridge = pd.read_csv(r'Data\\Features_Final\\greypartridge_features.csv')
features_goldfinch = pd.read_csv(r'Data\\Features_Final\\goldfinch_features.csv')
features_cornbunting = pd.read_csv(r'Data\\Features_Final\\cornbunting_features.csv')

features_dfs = [features_yellowhammer,
               features_whitethroat,
               features_turtledove,
               features_treesparrow,
               features_stockdove,
               features_starling,
               features_skylark,
               features_linnet,
               features_lapwing,
               features_greypartridge,
               features_goldfinch,
               features_cornbunting]
features = pd.concat(features_dfs, ignore_index=True)
```

Joining metadata and features DataFrames into one DataFrame named data by Recording ID. 

Recording ID is its own seperate column in the metadata DataFrame but is contained within a string in the clip column of features DataFrame, the recording ID therefore needs to be sliced from this clip column first. 

```python
lens = []
six, five = 0, 0
for v in metadata.Recording_ID:
    if len(str(v)) not in lens:
        lens.append(len(str(v)))
    if len(str(v)) == 6:
        six += 1
    if len(str(v)) == 5:
        five += 1

print(lens)
print(six, five) # Can see here that recording ID's are either 5 or 6 characters long
```

```python
import re

# This slices 6 characters from the clip column containing the recording ID.
# If chunk number is < 10 (single digit) then the slice is taken from -18 to -13 (inclusive)
# If chunk number is >= 10 (two digit) then the slice is taken from -19 to -14 (inclusive)
Recording_ID = []
for id in features['clip']:
    if re.search("chunk_...wav", id):
        Recording_ID.append(id[-19:-13])
    else:
        Recording_ID.append(id[-18:-12])
    
```

```python
# If Recording ID is 5 digits long then first character will be "-". If this is the case it will be removed
i = 0
for id in Recording_ID:
    if id[0] == '-':
        Recording_ID[i] = id[1:]
    i += 1
```

```python
#Recording ID column added to the features DataFrame
features['Recording_ID'] = Recording_ID
```

```python
features.Recording_ID = pd.to_numeric(features.Recording_ID)
```

```python
# Recording ID set as index for both metadata and features DataFrame, these will then be joined by index
metadata.set_index('Recording_ID', inplace=True)
features.set_index('Recording_ID', inplace=True)
```

```python
metadata.drop('Unnamed: 0', axis=1, inplace=True)
features.drop('Unnamed: 0', axis=1, inplace=True)
data = features.join(metadata, how="left")
```

```python
# New combined contains many columns which will not be useful for classification of birds. For example, the name of the person who uploaded the file.
# All columns will be examined here and irrelevant columns dropped from DataFrame. 
for col in data.columns:
    print(col)
```
